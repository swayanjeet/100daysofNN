```python
print(type(list_[0]))
list_ = list_[0:1000]
```

```python
from hmmlearn import hmm
model = hmm.GaussianHMM(n_components=80, n_iter=1000)
model.fit(np.array(list_).reshape(-1,1))
```

```python
output = model.sample(1000)
# print(counter[-1])
```

```python
list__ = output[0].tolist()
```

```python
counter = [i for i in range(0,len(list_))]
```
